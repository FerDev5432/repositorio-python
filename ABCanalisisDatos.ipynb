Transformacion de datos
* Para campos nuevos, calculados en base a otros campos
df["campoNuevo"] = df["campoAlQueSeLeAplicaLaFuncion"].apply(Funcion)


para cada valor del campo, aplicale la funcion y lo que retorne almacenalo en el campoNuevo


Funcion para calcular ganancias: toma el valor de los retweets y lo multiplica por un numero random entre 3 y 5

```python
import random
from numpy import random

def calcularGanancias(retweets):
  ganancia= retweets * random.randint(3,5)
  return ganancia

df["ganancias"] = df["retweets"].apply(calcularGanancias)
df.head()
```

```python
def popularidad(fila):
  resultado=fila["followees"]/fila["followers"]
  return resultado

df["popularidad"] = df.apply(popularidad, axis=1)
df.head()
```

Agrupacion de datos
* df.funcionGroupby("campo").funcionAgrupacion()

```python
df[['country', 'favorites', 'retweets', 'mentions', 'followers', 'followees', 'ganancias', 'popularidad']].groupby("country").mean()
```

* agrupacion de varios campos: agrupando el campo de country, aplica a las columnas diferentes metodos

```python
df.groupby("country").agg({
    "followers": 'sum',
    "mentions": 'mean',
    "retweets": 'max'
})
```
