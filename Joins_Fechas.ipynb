Joins - unir varios dataframes en uno solo

En este escenario tenemos 3 diferentes dataframes, pero buscamos la forma en como se relacionan todos tenemos columnas como usuario y ciudad que se encuentran como relaciones, a partir de aqui vamos a trabajar

```python
import pandas as pd
```

```python
#carga datos
df1 = pd.read_csv("dataset_2a.csv")
df2 = pd.read_csv("dataset_2b.csv")
df3 = pd.read_csv("dataset_2c.csv")
```

Para unir dataframes se hace como una analogia el dataframe principal sera el de la izquierda mientras que el de la derecha sera como un complemento

nombreDataframe1.merge(nombreDataframe2, on"CampoDeRelacion", how="tipoDeJoin")
en el tipo de join puede ir inner, left, right, dependiendo el caso de uso

```python
df_con_usuarios= df1.merge(df2, on='user', how="left")
df_con_usuarios
```

```python
df_completo = df_con_usuarios.merge(df3, on='ciudad', how="left")
df_completo
```

Generalmente las fechas estan como tipo de dato STR, pero lo ideal es tenerlas en formato de tipo fecha
dataframeNombre['campoFecha'] = pd.to_datetime(dataframeNombre['columna_fecha'],formatoFecha)

```python
df_completo['fecha'] = pd.to_datetime(df_completo['fecha_publicado'], format="%d/%m/%Y")
df_completo
```

Ejemplo: Quiero saber que dia de la semana es una fecha en especifico
Para cada valor de columnaConFormatoFecha se le va a aplicar la funcion strftime que convierte el formato fecha en un str y %A solo nos dara el dia.
nombreDataFrame[nombreColumna] = df['columnaConFormatoFecha'].apply(lambda x:x.strftime('%A'))


```python
df_completo['dia_semana'] = df_completo['fecha'].apply(lambda x:x.strftime('%A'))
df_completo
```

Ejemplo: Quiero saber los tweets por mes

```python
df_completo['mes'] = df_completo['fecha_publicado'].apply(lambda x:x[3:])
df_completo
```

```python
df_completo.groupby('mes').agg({
    'id':'count'
})
```

Agrupacion multiple: Agrupar x mes y x ciudad

* Nota: Aqui la unica columna que se reconoce del dataframe es id, mes y ciudad son indices, podemos dejar todo al mismo nivel
* ResentIndex, resetea el indice para que se convierta en una columna y asi todo tenerlo al mismo nivel, le debes pasar el nivel de indice 0,1,2,n,...

```python
agrupado = df_completo.groupby(['mes','ciudad']).agg({
    'id':'count'
})
agrupado
```

```python
desagrupado = agrupado.reset_index(1)
desagrupado
```

```python
desagrupado[['ciudad']]
```

Pivotar: es convertir cada valor unico de una columna en una nueva columna

```python
pivote = desagrupado.pivot(columns='ciudad')
pivote
```

Aqui eliminados el indice de id que en el Df anterior se mostrara arriba de las ciudades, de esta forma evitamos errores si queremos usar los indices de las ciudades

```python
pivote = desagrupado.pivot(columns='ciudad')
pivote = pivote.droplevel(0,axis='columns')
pivote
```

```python
pivote[['GUAYAQUIL']]
```

```python
pivote = pivote.fillna(0)
pivote
```

Transponer: Convertir columnas en filas y filas en columnas

```python
pivote.transpose()
```
