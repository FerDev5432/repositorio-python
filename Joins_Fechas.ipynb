Joins - unir varios dataframes en uno solo

En este escenario tenemos 3 diferentes dataframes, pero buscamos la forma en como se relacionan todos tenemos columnas como usuario y ciudad que se encuentran como relaciones, a partir de aqui vamos a trabajar

```python
import pandas as pd
```

```python
#carga datos
df1 = pd.read_csv("dataset_2a.csv")
df2 = pd.read_csv("dataset_2b.csv")
df3 = pd.read_csv("dataset_2c.csv")
```

Para unir dataframes se hace como una analogia el dataframe principal sera el de la izquierda mientras que el de la derecha sera como un complemento

nombreDataframe1.merge(nombreDataframe2, on"CampoDeRelacion", how="tipoDeJoin")
en el tipo de join puede ir inner, left, right, dependiendo el caso de uso

```python
df_con_usuarios= df1.merge(df2, on='user', how="left")
df_con_usuarios
```

```python
df_completo = df_con_usuarios.merge(df3, on='ciudad', how="left")
df_completo
```

Generalmente las fechas estan como tipo de dato STR, pero lo ideal es tenerlas en formato de tipo fecha

```python
df_completo['fecha'] = pd.to_datetime(df_completo['fecha_publicado'], format="%d/%m/%Y")
df_completo
```
